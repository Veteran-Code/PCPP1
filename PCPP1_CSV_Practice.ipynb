### Way Three With People - DictReader With Headers

```python
!head -n 1 /home/optimus/BreachedData/people.csv
```

```python
with open(f'{breached_data}people.csv', newline='') as csvfile:
    fieldnames = "Username,Company,Phone Number,UID,Zip Code,Email".split(',')
    
    people = csv.DictReader(csvfile, fieldnames=fieldnames)
    
    people_dataset = []
    
    # Be warned that this is special. 
    # It returns a dictionary with fieldnames as keys
    # Use the values method to change to a list
    for row in people:
        print(list(row.values()))
        people_dataset.append(list(row.values()))
```

```python
print(f'Be warned that this is special.\nIt returns a dictionary with fieldnames as keys.\nUse the values method to change to a list')
print('Also, note that the headers show up as the first row as well')
```

```python
print(f'Length should be 1000: {True if len(people_dataset[1:]) == 1000 else False}')
```

## Data Wrangling
- It appears that the people dataset and the credit card dataset is linked

```python
print(f'First entry in credit card dataset: {credit_card_dataset[1][0]}')
print(f'First entry in people dataset: {people_dataset[1][0]}')
print(f'The two datasets are of equal length: {True if len(credit_card_dataset) == len(people_dataset) else False}')
```

```python
print(f'Let\'s merge the two datasets together now!')

employee_dataset = {"headers": copy.deepcopy(people_dataset[0])}
employee_dataset["headers"].extend(copy.deepcopy(credit_card_dataset[0][1:]))

idx = 1 # Skip the headers

while idx < len(people_dataset):
    person = people_dataset[idx]
    credit_information = credit_card_dataset[idx]
    
    if person[0] != credit_information[0]:
        raise Exception(f'At index {idx}, the two datasets do not map')
    else:
        employee_dataset[person[0]] = copy.deepcopy(person[1:])
        employee_dataset[person[0]].extend(copy.deepcopy(credit_information[1:]))
    
    idx += 1
```

```python
print(f'Merged headers: {employee_dataset["headers"]}')
```

```python
print('First five rows of the employee dataset. Order may change across instances!')

count = 0

for key, val in employee_dataset.items():
    if key != 'headers':
        print(f'Name: {key}')
        print(f'Data: {val}')
        count += 1
        
    if count == 5:
        break
```
